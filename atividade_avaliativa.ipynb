```python
# Média do aluno

# Entrada

nota1 = float(input("Digite a nota do 1º bimestre: "))
nota2 = float(input("Digite a nota do 2º bimestre: "))
nota3 = float(input("Digite a nota do 3º bimestre: "))
nota4 = float(input("Digite a nota do 4º bimestre: "))

# Processamento

resultado = (nota1 + nota2 + nota3 + nota4) / 4

if resultado >= 7:
  print(f" A média da sua nota foi {resultado}, você foi aprovado!")

elif resultado >= 5:
   print(f" A média da sua nota foi {resultado}, você terá que fazer recuperação!")

else:

   print(f" A média da sua nota foi {resultado}, você foi reprovado!")
```

```python
# Positivo, negativo ou zero.

# Entrada

numero = int(input("Digite um número: "))

if numero > 0:
  print("O número é positivo")

elif numero < 0:
  print("O número é negativo")

else: # numero == 0
   print("Esse número é zero")
```

```python
# Calculadora simples, 04 operações.

print("--------------MENU-------------")
print("")
print("Opção - 1 = SOMA")
print("Opção - 2 = SUBTRÇÃO")
print("Opção - 3 = MULTIPLICAÇÃO")
print("Opção - 4 = DIVISÃO")

opcao = int (input("Digite a opção: "))

# Entrada
numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))

if opcao == 1:
  resultado = numero1 + numero2
  print(f" o resultado é {resultado}")

elif opcao == 2:
  resultado = numero1 - numero2
  print(f"o resultado é {resultado}")

elif opcao == 3:
  resultado = numero1 * numero2
  print(f"o resultado é {resultado}")

elif opcao == 4:
  resultado = numero1 / numero2
  print(f"o resultado é {resultado}")

else:
  print("Opção inválida")

```

```python
# Dia da semana

numero = int(input("Digite um número de 1 a 7"))

if numero == 1:
  print("Domingo")

elif numero == 2:
  print("Segunda-Feira")

elif numero == 3:
  print("Terça-Feira")

elif numero == 4:
  print("Quarta-Feira")


elif numero == 5:
  print("Quinta-Feira")

elif numero == 6:
  print("sexta-Feira")

elif numero == 7:
  print("Sábado-Feira")

else:
  print("Inválido")
```

```python
# Ano bissexto

numero = int(input("Digite um ano: "))

if (numero % 4 == 0 and numero % 100 != 0) or (numero % 400 == 0):

  print("Ano bissexto")

else:
  print("O ano não é bissexto")
```

```python
# Desconto em compras

valor = float(input("Digite o valor da compra: "))

if valor >= 100:
  desconto = 0.10

elif valor >= 50:
   desconto = 0.05

else:
  desconto = 0

valor_desconto = valor * desconto
total = valor - valor_desconto

print("Desconto aplicado", desconto)

print("Valor do desconto: ", valor_desconto)

print("Total a pagar", total)
```
