```python
# Operadores arítiméticos

# Fazer um programa que o usuário digita 02 números.
# Saída mostrar operacoes aritiméticas


# Entrada

a = int(input("digite um número:) "))
b = int(input("digite um número:) "))


# Processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b
resto = a % b

#Saída

print("")
print("-----------OPERADORES ARITIMÉTICOS------------")
print("                                              ")
print("---------------------SOMA---------------------")
print(f"O resultado da SOMA de {a} MAIS {b} é igual a : {soma}")
print("                                                       ")
print("------------------SUBTRAÇÃO---------------------")
print(f"O resultado da SUBTRAÇÃO de {a} MENOS {b} é igual a : {subtracao}")
print("                                                ")
print("-----------------MULTIPLICAÇÃO------------------")
print(f"O resultado da MULTIPLICAÇÃO de {a} VEZES {b} é igual a : {multiplicacao}")
print("                                                 ")
print("---------------------DIVISÃO---------------------")
print(f"O resultado da DIVISÃO de {a} DIVIDIDO {b} é igual a : {divisao}")
print("---------------------DIVISÃO---------------------")
print(f"O resultado da DIVISÃO INTEIRA de {a} DIVIDIDO {b} é igual a : {div_int}")
print("                                                 ")
print("-------------------------------------------------")
```
