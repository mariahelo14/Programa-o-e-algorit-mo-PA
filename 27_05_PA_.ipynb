```python
# Jogo menu inteligente

# Criar um sistema que funcione até o usuário escolher sair.

# Menu

# 1 - Par/Impar
# 2 - Positivo/ Negativo
# 3 - Maior
# 4 - Sair

opcao = 0

while opcao != 4:

    print("\n=====MENU=====")
    print("1 - Par/Ímpar")
    print("2 - Positivo/Negativo")
    print("3 - Maior")
    print("4 - Sair")


    try:
        opcao = int(input("Escolha uma opção: "))
    except ValueError:
        print("Entrada inválida. Por favor, digite um número.")
        continue # Continue to the next iteration of the loop

    # Par/Impar
    if opcao == 1:
       numero = int(input("Digite um número: "))

       if numero % 2 == 0:
         print("O número é PAR")

       else:
         print("O número é ÍMPAR")

    #2 Positivo/Negativo
    elif opcao == 2:
         numero = int(input("Digite um número: "))


         if numero > 0:
           print("Número POSITIVO")

         elif numero < 0:
           print("Número NEGATIVO")
         else:
           print("O número é ZERO")

    # 3 Maior Número
    elif opcao == 3:
      n1 = int(input("Digite o primeiro número: "))
      n2 = int(input("Digite o segundo número: "))

      if n1 > n2:
        print("O maior número é:", n1)

      elif n2 > n1:
        print("O número maior é:", n2)

      else:
        print("Os números são iguais!")

    elif opcao == 4:
        print("Programa encerrado!")


    else:
      print("Opção inválida, tente novamente!")
```

```python
# Urna Eletônica

# Criar um programa de votação com 3 candidatos

# Votar candidato A
# Votar candidato B
# Votar candidato C
# Votar Branco
# Votar Nulo
# Encerrar votação

# o sistema deve repetir até encerrar a votação
# Contar votos de cada candidato
# contar votos brancos e nulos
# ao encerrar, mostrar:

# total de votos
# total de votospor candidatos
# total de votos brancos
# total de votos nulos
# vencedor da eleição

# regras extras
# se ouver empate entre  candidatos : Mostrar eleição empatada
# se não houver votos válidos, mostrar : "Nenhum candidato recebeu votos."


a = 0
b = 0
c = 0
branco = 0
nulo = 0

opcao = 0

while opcao != 6:

    print("\n===== URNA ELETRÔNICA === ==")
    print("1 - Votar candidato A")
    print("2 - Votar candidato B")
    print("3 - Votar candidato C")
    print("4 - Voto Branco")
    print("5 - Voto Nulo")
    print("6 - Encerrar votação")

    try:
        opcao = int(input("Escolha uma opção: "))
    except ValueError:
        print("Entrada inválida. Por favor, digite um número.")
        continue

    if opcao == 1:
        a += 1
        print("Voto no candidato A registrado!")

    elif opcao == 2:
        b += 1
        print("Voto no candidato B registrado!")

    elif opcao == 3:
        c += 1
        print("Voto no candidato C registrado!")

    elif opcao == 4:
        branco += 1
        print("Voto BRANCO registrado!")

    elif opcao == 5:
        nulo += 1
        print("Voto NULO registrado!")

    elif opcao == 6:
        print("\nEncerrando votação...")

# totais
total = a + b + c + branco + nulo
validos = a + b + c

print("\n===== RESULTADO FINAL === ==")
print("Total de votos:", total)

print("\nVotos por candidato:")
print("Candidato A:", a)
print("Candidato B:", b)
print("Candidato C:", c)

print("\nVotos brancos:", branco)
print("Votos nulos:", nulo)

# verificar vencedor
if validos == 0:
    print("\nNenhum candidato recebeu votos.")

elif a > b and a > c:
    print("\nVencedor: Candidato A")

elif b > a and b > c:
    print("\nVencedor: Candidato B")

elif c > a and c > b:
    print("\nVencedor: Candidato C")

else:
    print("\nEleição empatada!")
```
