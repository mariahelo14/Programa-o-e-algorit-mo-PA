```python
# Escrever um programa que o usuario digita uma palavra e o programa verifica se a palavra é uma palindroma
palavra = input("Digite uma palavra: ")
invertida = ""

for i in (palavra):
   invertida = i + invertida

if palavra == invertida:
   print (f"{palavra} é uma palavra palindroma.")

else:
  print (f"{palavra} não é uma palavra palindroma.")
```

```python
# Soma acumulada
# Peça ao usuário para digitar um numero e imprimi o resultado da soma 1 até o número digitado.

numero = int(input("Digite um número: "))

if numero < 0:
  print("Digite um número positivo.")

else:
  soma_acumulada = 0
  for i in range (1, numero + 1):
      soma_acumulada += i

  print (f"A soma acumulada de 1 até {numero} é: {soma_acumulada}")
```

```python
# Peça um número ao usuário e o programa irá separar números pares e impares
# Exemplo de saída:

# Pares: 0, 2, 4, 6 ...
# Impares: 1, 3, 5, 7 ...

numero = int(input("Digite um número para separar pares e ímpares: "))

pares_str ="Pares"
impares_str = "Ímpares"

primeiro_par = True
primeiro_impar = True

for i in range(numero + 1):
    if i % 2 == 0:
        if primeiro_par:
            pares_str += f": {i}"
            primeiro_par = False
        else:
            pares_str += f", {i}"
    else:
        if primeiro_impar:
            impares_str += f": {i}"
            primeiro_impar = False
        else:
            impares_str += f", {i}"

print(pares_str)
print(impares_str)
```
