```python
# 1 - Contagem simples, o usuario digita um numero e o programa imprimi do zero até o número.

numero = int(input("Digite um número: "))

for i in range (0, numero + 1):
   print(i)
```

```python
# 2 - Contagem regressiva, o usuário digita um número e programa imprimi do número até o zero.

numero = int(input("Digite um número para mostrar sua ordem regressiva: "))

for i in range (numero, -1, -1):
    print(i)
```

```python
# 3 - Números pares, o usuário digita um número e o programa imprimi números pares.

numero = int(input("Digite um número para separar os números pares: "))

pares_str = ""
primeiro_par = True

for i in range(numero + 1):
    if i % 2 == 0:
        if primeiro_par:
            pares_str += f"{i}"
            primeiro_par = False
        else:
            pares_str += f", {i}"
print(f"Pares: {pares_str}")
```

```python
# 4 - Números impares, o usuário digita um número e o programa imprimi os números impares.

numero = int(input("Digite um número para mostrar os números impares: "))

for i in range(1, numero + 1, 2):
    print(i)
```

```python
# 5 - Soma, o usuário digita um número e o programa soma os números do zero até o número.

numero = int(input("Digite um número: "))

if numero < 0:
  print("Digite um número positivo.")

else:
  soma_acumulada = 0
  for i in range (0, numero + 1):
      soma_acumulada += i

  print (f"A soma de 0 até {numero} é: {soma_acumulada}")
```

```python
# 6 - Tabuada, o usuário digita um número e o programa imprimi a tabuada do número.

numero = int(input("Digite um número e descubra sua tabuada: "))

print(f"\nTabuada do número {numero}: \n")

for i in range (1,11):
  resultado = numero * i
  print(f"{numero} x {i} = {resultado}")
```
